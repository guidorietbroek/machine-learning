```python
customers = pd.read_excel('klantenbestand_03112021.xlsx')
```

```python
customers.head()
```

```python
customers.info()
```

```python
customers = customers.rename(columns = {"Debiteurnummer" : "debnr"})
```

```python
customers.head()
```

```python
customers['debnr'] = pd.to_numeric(customers['debnr'],errors='coerce')
```

```python
customers['debnr'] = customers['debnr'].fillna(0)
```

```python
customers['debnr'] = customers['debnr'].astype(int)
```

```python
customers.head()
```

<h2>Joinen van data</h2>

```python
df = orders.merge(right=customers, left_on='debnr', right_on='debnr', how='left')
```
